```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.9,
                       random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.xlabel('')
plt.show()
```

```python
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
```

```python
gm = GaussianMixture(n_components=4)
gm.fit(X)
gm_cluster_labels = gm.predict(X)
```

```python
scatter = plt.scatter(X[:,0], X[:,1], c=gm_cluster_labels)
```

```python
#real data
scatter = plt.scatter(X[:,0], X[:,1], c=y_true)
```

```python
# K Means
kmean = KMeans(n_clusters=4)
kmean.fit(X)
kmean_cluster_label = kmean.predict(X)
scatter = plt.scatter(X[:,0], X[:,1], c=kmean_cluster_label)
```

```python
#DBscan
db = DBSCAN(eps=0.7, min_samples=15)
db.fit(X)
db_labels = db.fit_predict(X)
scatter = plt.scatter(X[:,0], X[:,1], c=db_labels)
```

```python
#Agglomerative
ac = AgglomerativeClustering(n_clusters=4)
ac.fit(X)
ac_label = ac.fit_predict(X)
scatter = plt.scatter(X[:,0], X[:,1], c=ac_label)
```

# Mall Customer Datasets

```python
data = pd.read_csv('Mall_Customers.csv')
data.head()
```

```python
data = data.iloc[:,3:5]
data.head()
```

```python
scatter = plt.scatter(data.values[:,0], data.values[:,1])
```

```python
model = KMeans(n_clusters=5)

model.fit(data)
labels= model.predict(data)
plt.scatter(data.values[:,0], data.values[:,1], c=labels)
plt.show()
```

```python
#metode elbow
variance = []

for i in range(1,10):
  kmean = KMeans(n_clusters = i, max_iter=300, n_init=10, random_state=0)
  kmean.fit(data)
  variance.append(kmean.inertia_)
```

```python
plt.plot(range(1, 10), variance)
plt.title("The Elbow Method")
plt.show()
```
